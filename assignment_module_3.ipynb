## Tidy GDP

Manipulate the GDP.csv file and make it tidy, the result should be a pandas dataframe with the following columns:
* Country Name
* Country Code
* Year
* GDP

```python
GDP_df = pd.df = pd.read_csv("https://raw.githubusercontent.com/gumdropsteve/intro_to_python/main/day_09/data/gdp.csv")
GDP_df.head(2)
```

```python
GDP_df.columns[4:62]
```

```python
GDP_df1 = GDP_df.drop(['Indicator Name', 'Indicator Code', 'Unnamed: 62'], axis =1)
GDP_df1
```

```python
GDP_df1 = GDP_df1.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='GDP')
GDP_df1.head()
```

```python
GDP_df1.isnull().sum()
```
