```
from sklearn.feature_extraction import DictVectorizer

vec = DictVectorizer(sparse=False, dtype=int)

# get list of dicts from `type` column
data = [{'type':v} for v in df.type]

d = vec.fit_transform(data)

type_df = pd.DataFrame(d, columns=vec.feature_names_)
type_df.tail(3)
```

```
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer()
X = vec.fit_transform(df.title)

title_df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

```
title_df
```

```
import numpy as np

np.sum(pd.DataFrame(X.toarray(), columns=vec.get_feature_names()))
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
sample = df['title']

vec = TfidfVectorizer()

X = vec.fit_transform(sample)

Tfid_df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

```
Tfid_df
```

```
big_df = pd.concat([df, type_df, Tfid_df], axis=1)

big_df
```
